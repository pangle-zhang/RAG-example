```python
import chromadb
import os, sys, json
CURRENT_DIR = os.getcwd()
sys.path.append(os.path.join(CURRENT_DIR, '../'))
from utils.embed import embed

question = "天府新区哪里可以挖折耳根？"
embedding = embed(question)

os.environ["NO_PROXY"] = "*"
chroma_db = chromadb.HttpClient(
    host="localhost", 
    port=8082, 
    ssl=False
)
collection = chroma_db.get_collection('rag_example')
results = collection.query(embedding, n_results=3)
print(f"Question: {question}")
print(f"Total results: {len(results)}")
print(json.dumps(results, indent=2, ensure_ascii=False))
print(json.dumps(results["documents"], indent=2, ensure_ascii=False))
```
